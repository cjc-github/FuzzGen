````python
def get_return_parm(one_func,project):
    stack = [one_func.node]
    analyze_list = []
    while stack:
        node = stack.pop()
        # 在这里处理节点的逻辑，比如打印节点的类型、内容等
        # print("Node Type:", node.type)
        # print("Node Content:", node.utf8_content)
        # 将子节点添加到堆栈中，以便迭代处理
        for child in node.children:
            if child.type == "function_declarator":
                continue
            elif child.type == "compound_statement":
                continue
            stack.append(child)
            analyze_list.append(child)
    return_list = []
    for node in analyze_list:
        if node.text in project.all_class_dict:
            return_list.append(project.all_class_dict[node.text])
    return_list = list(set(return_list))
    if len(return_list) == 1:
        return return_list[0]
    elif len(return_list) > 1:
        print("ERROR",return_list)
        return return_list
    else:
        return return_list
````
